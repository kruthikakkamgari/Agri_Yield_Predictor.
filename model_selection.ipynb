```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor 
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np
import pandas as pd

# ---------------- Load Features & Target ----------------
X_loe_mm, y_loe_mm = df_loe_mm.drop(columns=["Yield"]), df_loe_mm["Yield"]

# ------ Random Forest (Squared Error) ------
rf_se = RandomForestRegressor(n_estimators=100, random_state=42, criterion="squared_error")
rf_se.fit(X_loe_mm, y_loe_mm)
y_pred_se = rf_se.predict(X_loe_mm)

rf_se_row = [
    "Random Forest (Squared Error)",
    round(r2_score(y_loe_mm, y_pred_se), 4),
    round(mean_squared_error(y_loe_mm, y_pred_se), 4),
    round(mean_absolute_error(y_loe_mm, y_pred_se), 4),
    round(np.sqrt(mean_squared_error(y_loe_mm, y_pred_se)), 4)
]

# ------ Random Forest (Absolute Error) ------
rf_ae = RandomForestRegressor(n_estimators=100, random_state=42, criterion="absolute_error")
rf_ae.fit(X_loe_mm, y_loe_mm)
y_pred_ae = rf_ae.predict(X_loe_mm)

rf_ae_row = [
    "Random Forest (Absolute Error)",
    round(r2_score(y_loe_mm, y_pred_ae), 4),
    round(mean_squared_error(y_loe_mm, y_pred_ae), 4),
    round(mean_absolute_error(y_loe_mm, y_pred_ae), 4),
    round(np.sqrt(mean_squared_error(y_loe_mm, y_pred_ae)), 4)
]

# ------ Decision Tree (Squared Error) ------
dt_se = DecisionTreeRegressor(random_state=42, criterion="squared_error", max_depth=None)
dt_se.fit(X_loe_mm, y_loe_mm)
y_pred_dt = dt_se.predict(X_loe_mm)

dt_se_row = [
    "Decision Tree (Squared Error)",
    round(r2_score(y_loe_mm, y_pred_dt), 4),
    round(mean_squared_error(y_loe_mm, y_pred_dt), 4),
    round(mean_absolute_error(y_loe_mm, y_pred_dt), 4),
    round(np.sqrt(mean_squared_error(y_loe_mm, y_pred_dt)), 4)
]

# ------ Decision Tree (Absolute Error) ------
dt_ae = DecisionTreeRegressor(random_state=42, criterion="absolute_error", max_depth=None)
dt_ae.fit(X_loe_mm, y_loe_mm)
y_pred_dt_ae = dt_ae.predict(X_loe_mm)

dt_ae_row = [
    "Decision Tree (Absolute Error)",
    round(r2_score(y_loe_mm, y_pred_dt_ae), 4),
    round(mean_squared_error(y_loe_mm, y_pred_dt_ae), 4),
    round(mean_absolute_error(y_loe_mm, y_pred_dt_ae), 4),
    round(np.sqrt(mean_squared_error(y_loe_mm, y_pred_dt_ae)), 4)
]

# ------ XGBoost ------
xgb_model = XGBRegressor(random_state=42)
xgb_model.fit(X_loe_mm, y_loe_mm)
y_pred_xgb = xgb_model.predict(X_loe_mm)

xgb_row = [
    "XGBoost",
    round(r2_score(y_loe_mm, y_pred_xgb), 4),
    round(mean_squared_error(y_loe_mm, y_pred_xgb), 4),
    round(mean_absolute_error(y_loe_mm, y_pred_xgb), 4),
    round(np.sqrt(mean_squared_error(y_loe_mm, y_pred_xgb)), 4)
]

# ------ Support Vector Regressor (SVM) ------
svm_model = SVR(kernel="rbf")
svm_model.fit(X_loe_mm, y_loe_mm)
y_pred_svm = svm_model.predict(X_loe_mm)

svm_row = [
    "Support Vector Regressor (SVM)",
    round(r2_score(y_loe_mm, y_pred_svm), 4),
    round(mean_squared_error(y_loe_mm, y_pred_svm), 4),
    round(mean_absolute_error(y_loe_mm, y_pred_svm), 4),
    round(np.sqrt(mean_squared_error(y_loe_mm, y_pred_svm)), 4)
]

# ------ Gradient Boosting Regressor ------
gbr_model = GradientBoostingRegressor(random_state=42)
gbr_model.fit(X_loe_mm, y_loe_mm)
y_pred_gbr = gbr_model.predict(X_loe_mm)

gbr_row = [
    "Gradient Boosting Regressor",
    round(r2_score(y_loe_mm, y_pred_gbr), 4),
    round(mean_squared_error(y_loe_mm, y_pred_gbr), 4),
    round(mean_absolute_error(y_loe_mm, y_pred_gbr), 4),
    round(np.sqrt(mean_squared_error(y_loe_mm, y_pred_gbr)), 4)
]

# ------ AdaBoost Regressor ------
ada_model = AdaBoostRegressor(random_state=42)
ada_model.fit(X_loe_mm, y_loe_mm)
y_pred_ada = ada_model.predict(X_loe_mm)

ada_row = [
    "AdaBoost Regressor",
    round(r2_score(y_loe_mm, y_pred_ada), 4),
    round(mean_squared_error(y_loe_mm, y_pred_ada), 4),
    round(mean_absolute_error(y_loe_mm, y_pred_ada), 4),
    round(np.sqrt(mean_squared_error(y_loe_mm, y_pred_ada)), 4)
]

# ------ Linear Regression ------
lin_model = LinearRegression()
lin_model.fit(X_loe_mm, y_loe_mm)
y_pred_lin = lin_model.predict(X_loe_mm)

lin_row = [
    "Linear Regression",
    round(r2_score(y_loe_mm, y_pred_lin), 4),
    round(mean_squared_error(y_loe_mm, y_pred_lin), 4),
    round(mean_absolute_error(y_loe_mm, y_pred_lin), 4),
    round(np.sqrt(mean_squared_error(y_loe_mm, y_pred_lin)), 4)
]

# ------ Ridge Regression ------
ridge_model = Ridge(random_state=42)
ridge_model.fit(X_loe_mm, y_loe_mm)
y_pred_ridge = ridge_model.predict(X_loe_mm)

ridge_row = [
    "Ridge Regression",
    round(r2_score(y_loe_mm, y_pred_ridge), 4),
    round(mean_squared_error(y_loe_mm, y_pred_ridge), 4),
    round(mean_absolute_error(y_loe_mm, y_pred_ridge), 4),
    round(np.sqrt(mean_squared_error(y_loe_mm, y_pred_ridge)), 4)
]

# ------ Lasso Regression ------
lasso_model = Lasso(random_state=42)
lasso_model.fit(X_loe_mm, y_loe_mm)
y_pred_lasso = lasso_model.predict(X_loe_mm)

lasso_row = [
    "Lasso Regression",
    round(r2_score(y_loe_mm, y_pred_lasso), 4),
    round(mean_squared_error(y_loe_mm, y_pred_lasso), 4),
    round(mean_absolute_error(y_loe_mm, y_pred_lasso), 4),
    round(np.sqrt(mean_squared_error(y_loe_mm, y_pred_lasso)), 4)
]

# ------ ElasticNet Regression ------
en_model = ElasticNet(random_state=42)
en_model.fit(X_loe_mm, y_loe_mm)
y_pred_en = en_model.predict(X_loe_mm)

en_row = [
    "ElasticNet Regression",
    round(r2_score(y_loe_mm, y_pred_en), 4),
    round(mean_squared_error(y_loe_mm, y_pred_en), 4),
    round(mean_absolute_error(y_loe_mm, y_pred_en), 4),
    round(np.sqrt(mean_squared_error(y_loe_mm, y_pred_en)), 4)
]

# ------ KNN Regressor ------
knn_model = KNeighborsRegressor(n_neighbors=5)
knn_model.fit(X_loe_mm, y_loe_mm)
y_pred_knn = knn_model.predict(X_loe_mm)

knn_row = [
    "KNN Regressor",
    round(r2_score(y_loe_mm, y_pred_knn), 4),
    round(mean_squared_error(y_loe_mm, y_pred_knn), 4),
    round(mean_absolute_error(y_loe_mm, y_pred_knn), 4),
    round(np.sqrt(mean_squared_error(y_loe_mm, y_pred_knn)), 4)
]

# ------ Final Table ------
df_results = pd.DataFrame(
    [
        rf_se_row, rf_ae_row,
        dt_se_row, dt_ae_row,
        xgb_row, svm_row,
        gbr_row, ada_row,
        lin_row, ridge_row,
        lasso_row, en_row,
        knn_row
    ],
    columns=["Model", "R²", "MSE", "MAE", "RMSE"]
)

print(df_results.to_string(index=False))
```
